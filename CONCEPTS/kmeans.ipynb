# Create Sample Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create synthetic data
np.random.seed(42)

X1 = np.random.randn(100, 2) + np.array([2, 2])
X2 = np.random.randn(100, 2) + np.array([-2, -2])
X3 = np.random.randn(100, 2) + np.array([2, -2])

X = np.vstack([X1, X2, X3])
```

# Visualize Raw Data (Before Clustering)

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.title("Raw Data (No Labels)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# Apply K-Means

```python
kmeans = KMeans(
    n_clusters=3,
    n_init=10,
    random_state=42
)

kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

# Visualize Clusters + Centroids

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    s=200,
    marker='X'
)
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
costs = []

for k in range(1, 8):
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(X)
    costs.append(km.inertia_)

plt.figure()
plt.plot(range(1, 8), costs, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Cost J")
plt.title("Elbow Method")
plt.show()
```
