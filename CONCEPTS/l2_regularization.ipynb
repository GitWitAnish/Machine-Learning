```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(2,)),
    tf.keras.layers.Dense(
        16,
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.01)
    ),
    tf.keras.layers.Dense(
        3,
        activation='softmax',
        kernel_regularizer=tf.keras.regularizers.l2(0.01)
    )
])
```
