Normalization


```python
import numpy as np

# Example dataset (each column = a feature)
X = np.array([[2104, 3],
              [1600, 3],
              [2400, 3],
              [1416, 2],
              [3000, 4]])


mu = np.mean(X, axis=0)
sigma = np.std(X, axis=0)


X_norm = (X - mu) / sigma


print("Normalized features:\n", X_norm)
```
